### Does Household Income Affect childrens Outcomes? A Systematic Review of the Evidence 
**Kerris Cooper & Kitty Stewart** (2021)

household income has a positive causal effect on childrens outcomes, including their cognitive and social-behavioural<br> development and their health, particularly in households with low income to begin with. There is also clear evidence of a<br> positive causal effect of income on intermediate outcomes that are important for childrens development, including maternal<br> mental health, parenting and the home environment.


**Investment theory** captures the way financial resources enable parents to buy goods that children need in order to thrive:<br> good quality housing and a healthy diet; books and other learning materials; outings and holidays (Duncan et al. 2017)<br>

**The family stress model** refers to the impact of economic disadvantage on the emotional home environment, holding that hardship causes stress for parents, and this in turn affects their parenting abilities (Conger et al. 2000).

```python
Dm3_df = Domain3.drop(['child_age','child_height'],axis = 1)
Dm3_df
```

```python
#np.nan_to_num(Domain3)
Dm3_df1 = pd.DataFrame(Dm3_df).fillna(Dm3_df.mean()) #replace null values with mean
```

```python
Dm3_df1.isna().sum()
```

```python
X = Dm3_df1.drop('child_domain_3', axis=1)
y = Dm3_df1['child_domain_3']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# create scaler object
scaler = StandardScaler()
```

```python
# create scaled version of the predictors (there is no need to scale the response)
X_scaled = scaler.fit_transform(X)
```

```python
# convert the scaled predictor values into a dataframe
X_standardise = pd.DataFrame(X_scaled,columns=X.columns)
X_standardise.head()
```

```python
# Import train/test splitting function from sklearn
from sklearn.model_selection import train_test_split
```

```python
# Split the data into train and test, being sure to use the standardised predictors
X_train, X_test, y_train, y_test = train_test_split(X_standardise, 
                                                    y, 
                                                    test_size=0.2, 
                                                    shuffle=False)
```

```python
# Import the ridge regression module from sklearn
from sklearn.linear_model import Ridge
```

```python
# Create ridge model
ridge = Ridge()
```

```python
# Train the model
ridge.fit(X_train, y_train)
```

```python
# Extract the model intercept value
b0 = float(ridge.intercept_)
```

```python
# Extract the model coefficient value
coeff = pd.DataFrame(ridge.coef_, X.columns, columns=['Coefficient'])
```

```python
print("Intercept:", float(b0))
```

```python
# Check out the coefficients
coeff
```

```python
# Fit a basic linear model
from sklearn.linear_model import LinearRegression

# Create model object
lm = LinearRegression()

# Train model
lm.fit(X_train, y_train)
```

```python
# Import metrics module
from sklearn import metrics
```

```python
# Check training accuracy
train_lm = lm.predict(X_train)
train_ridge = ridge.predict(X_train)

print('Training MSE')
print('Linear:', metrics.mean_squared_error(y_train, train_lm))
print('Ridge :', metrics.mean_squared_error(y_train, train_ridge))
```

```python
test_lm = lm.predict(X_test)
test_ridge = ridge.predict(X_test)

print('Testing MSE')
print('Linear:', metrics.mean_squared_error(y_test, test_lm))
print('Ridge :', metrics.mean_squared_error(y_test, test_ridge))
```

```python
a = Obs_df['child_observe_attentive'].value_counts()
b = Obs_df['child_observe_concentrated'].value_counts()
c = Obs_df['child_observe_diligent'].value_counts()
d = Obs_df['child_observe_interested'].value_counts()

print(a)
print(b)
print(c)
print(d)
```

```python
#Coding the variables using a function
def my_recode(column):
    #df = Obs_df
    if column == "Often":
        return 1
    elif column == "Sometimes":
        return 2
    elif column == "Almost always":
        return 3
    else:
        column == "Almost never"
        return 0
```

```python
Obs_df.columns
```

```python
my_recode('Often'), my_recode('Sometimes'), my_recode('Almost always'), my_recode('Almost never')
```

```python
#Apply my_code function on Obs_df columns (categorical)
Obs_df['child_observe_attentive_D'] = Obs_df['child_observe_attentive'].apply(my_recode)
Obs_df['child_observe_concentrated_D'] = Obs_df['child_observe_concentrated'].apply(my_recode)
Obs_df['child_observe_diligent_D'] = Obs_df['child_observe_diligent'].apply(my_recode)
Obs_df['child_observe_interested_D'] = Obs_df['child_observe_interested'].apply(my_recode)
```
